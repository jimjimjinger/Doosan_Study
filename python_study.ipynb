### 문제 1. safe_divide 함수 만들기

- b가 0이면 메시지 출력 후 None 반환
- 아니면 a / b 반환

````python
def safe_divide(a, b):
    if b == 0:
        return None
    else:
        return a/b

result1 = safe_divide(10, 2)
print(result1)

result2 = safe_divide(10, 0)
print(result2)
````

### 문제 3. swap_return 함수 만들기

#### 요구사항
- 매개변수 am b에 전달된 두 값을 서로 교환해야 한다.
- 교환된 두 갑은 튜플 형태로 return해야 한다

#### 실행 결과
20 10

````python
def swap_return(a, b):
    temp = a
    a = b
    b = temp
    return a, b

x = 10
y = 20
x, y = swap_return(x, y)
print(x, y)
````

### 문제 1. (클래스/인스턴스 변수) 회원 수 카운터 만들기

`Member` 클래스를 작성하시오.

#### 요구사항
- 클래스 변수 `count` 를 만들어 현재까지 생성된 객체 수를 저장 (처음 0)
- `__init__(self, name)` 가 호출될 때마다
  - 인스턴스 변수 `self.name` 저장
  - `Member.count` 를 1 증가
- `info(self)` 메서드는 `"name: {이름}"` 을 출력

````python

class Member:
    count = 0
    def __init__(self, name):
        self.name = name
        Member.count += 1
    def info(self):
        print(f"name: {self.name}")


m1 = Member("Jimin")
m2 = Member("Minji")
m1.info()                 # name: Jimin
m2.info()                 # name: Minji
print(Member.count)       # 2
````

### 문제 2. (리스트 멤버변수) 값 관리 클래스 만들기

`ValueBox` 클래스를 작성하시오.

#### 요구사항
- `__init__`에서 `self.data` 를 빈 리스트로 초기화
- `add(self, num)` : num을 `self.data`에 추가
- `fprint(self)` : 리스트 그대로 출력
- `plus(self)` : 현재 리스트의 모든 값을 더한 합을 **return**

````python
class ValueBox:
    def __init__(self):
        self.data = []
    def add(self, num):
        self.data.append(num)
    def fprint(self):
        print(self.data)
    def plus(self):
        return sum(self.data)

box = ValueBox()
box.add(10)
box.add(20)
box.add(30)
box.fprint()          # [10, 20, 30]
print(box.plus())     # 60
````

### 문제 4. (상속 + 오버라이딩 + super) 소개 메서드 확장하기

`Human` 클래스를 상속받는 `Student` 클래스를 작성하시오.

#### 요구사항
- `Human.__init__(self, name, age)` : 이름/나이 저장
- `Human.introduce(self)` : 아래 2줄 출력
  - `이름:{name}`
  - `나이:{age}`
- `Student(Human)`
  - `__init__(self, name, age, studentNum)` 에서 `super().__init__(name, age)` 사용 후 학번 저장
  - `introduce(self)` 를 오버라이딩해서
    1) `super().introduce()` 실행
    2) 그 다음 줄에 `학번:{studentNum}` 출력

````python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def introduce(self):
        print(f"이름: {self.name}")
        print(f"나이: {self.age}")

class Student(Human):
    def __init__(self, name ,age, studentNum):
        super().__init__(name, age)
        self. studentNum = studentNum
    def introduce(self):
        super().introduce()
        print(f"학번: {self.studentNum}")

s = Student("홍길동", 20, 2025001)
s.introduce()
````

### 문제 2) 점수 리스트 검증기 (raise + 예외 메시지)

점수 리스트를 입력 받아 **0~100 사이의 정수만** 허용하는 검증 함수를 작성하세요.

#### 요구사항
1. 함수 이름: `validate_scores(scores)`
2. 매개변수 `scores`는 리스트이며, 각 원소는 정수여야 한다.
3. 다음 조건 중 하나라도 만족하면 `raise`로 예외를 발생시킨다.
   - 원소가 정수가 아니면: `TypeError("정수만 허용")`
   - 점수가 0~100 범위를 벗어나면: `ValueError("범위 오류")`
4. 모든 점수가 정상이라면 평균을 `return`한다.
5. 아래 테스트 코드가 동작해야 한다.

#### 테스트 코드(그대로 사용)
```python
tests = [
    [100, 90, 80],
    [100, -1, 80],
    [100, 'A', 80],
]

for t in tests:
    try:
        avg = validate_scores(t)
        print('평균:', avg)
    except Exception as e:
        print('에러:', e)
```

#### 기대 출력(형식)
- 첫 번째는 평균 출력
- 두 번째는 `에러: 범위 오류`
- 세 번째는 `에러: 정수만 허용`

````python
def validate_scores(scores):
    try:
        print(sum(scores)/len(scores))
        for i in scores:
            if i < 0 or i > 100:
                raise ValueError
    except ValueError:
        print("범위를 벗어났습니다.")
    except TypeError:
        print("정수를 입력하시오")

tests = [
    [100, 90, 80],
    [100, -1, 80],
    [100, 'A', 80],
]

for t in tests:
    try:
        avg = validate_scores(t)
        print('평균:', avg)
    except Exception as e:
        print('에러:', e)
````
