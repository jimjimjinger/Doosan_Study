# 스터디 1주차 보고서

### GitHub 연동 및 보고서 업로드

이번 스터디에서는 GitHub를 활용하여 스터디 내용을 정리하고 공유하는 과정을 직접 진행하였다.
먼저 공용 GitHub Repository를 생성하고, 모든 팀원이 접근할 수 있도록 권한을 설정하였다.
이를 통해 각자 작성한 스터디 자료를 한 공간에서 관리할 수 있는 환경을 만들었다.

보고서 작성 형식으로는 Jupyter Notebook(.ipynb)을 사용하기로 하였다.
처음에는 ipynb 사용이 익숙하지 않아 글 작성 방식에 대해 간단히 서로 공유하는 시간을 가졌다.
Markdown 셀과 코드 셀의 차이를 확인하고, 실제로 간단한 글과 코드를 작성해 보았다.

각자 VS Code에서 ipynb 파일을 열어 스터디 내용 정리를 시도하였고,
작성한 파일을 GitHub에 커밋하고 푸시하는 과정까지 함께 진행하였다.
이 과정에서 GitHub 연동과 업로드 흐름을 자연스럽게 익힐 수 있었다.

이번 활동을 통해 스터디 내용을 단순히 말로 정리하는 것이 아니라,
문서와 코드 형태로 남기는 방법의 중요성을 체감할 수 있었다.
앞으로는 매 회차 스터디 내용을 동일한 방식으로 정리해 나갈 예정이다.
